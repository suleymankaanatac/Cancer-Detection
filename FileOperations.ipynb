```python
import csv
from PIL import Image
import cv2
import os
import shutil
import PIL

file=open('train.csv')
type(file)
csvreader=csv.reader(file)
header=[]
header=next(csvreader)
normal=[]
cancer=[]

for row in csvreader:
    if int(row[1])==0:
        normal.append(row[0]+".jpeg")
    else:
        cancer.append(row[0]+".jpeg")

source=r'C:\Users\KAAN\Desktop\cancer detection\part1\images'
dest1=r'C:\Users\KAAN\Desktop\cancer detection\part1\cancerno'
dest2=r'C:\Users\KAAN\Desktop\cancer detection\part1\cancer'

files=os.listdir(source)
for f in files:
    print(f)
    if f in normal:

        src_path = os.path.join(source, f)
        dst_path = os.path.join(dest1, f)
        shutil.move(src_path, dst_path)
    elif f in cancer:
        src_path = os.path.join(source, f)
        dst_path = os.path.join(dest2, f)
        shutil.move(src_path, dst_path)
```
