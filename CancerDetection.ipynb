```python
import numpy as np
import cv2
import os
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense,Dropout
from sklearn.metrics import confusion_matrix
from sklearn.metrics import precision_score, recall_score, f1_score
```

```python
labels=[]
data=[]

ccount=0
ncount=0
tcount=0

cancerImages=os.listdir('cancer/')
for imageNames in cancerImages:
    
    img=cv2.imread('cancer/'+imageNames)
    img = cv2.resize(img, (200, 200))
    data.append(np.array(img))
    labels.append(1)
    ccount+=1

normalImages=os.listdir('cancerno/')
for imageNames in normalImages:
   
    img=cv2.imread('cancerno/'+imageNames)
    img = cv2.resize(img, (200, 200))
    data.append(np.array(img))
    labels.append(0)
    ncount+=1
testData=[]        


testImages=os.listdir('images/')
for imageNames in testImages:
    
    img=cv2.imread('images/'+imageNames)
    img = cv2.resize(img, (200, 200))
    testData.append(np.array(img))
    tcount+=1

print(ccount)
print(ncount)
print(tcount)
#saving them as np arrays so that I can make operations with them
data = np.array(data)
testData = np.array(testData)
labels = np.array(labels)
        
```

```python
#splitting dataset into train and test sets
X_train, X_val, y_train, y_val = train_test_split(data, labels, test_size=0.2, random_state=42)
```

```python
X_train = X_train / 255.0
X_val = X_val / 255.0
testData=testData / 255.0
```

```python
model = Sequential()
model.add(Conv2D(32, (3, 3), activation='relu', input_shape=(200, 200, 3)))
model.add(MaxPooling2D((2, 2)))
model.add(Dropout(0.25))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D((2, 2)))
model.add(Dropout(0.25)) 
model.add(Conv2D(128, (3, 3), activation='relu'))
model.add(MaxPooling2D((2, 2)))
model.add(Dropout(0.25))  
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5)) 
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=25, batch_size=32, validation_data=(X_val, y_val))
```

```python
train_preds=model.predict(X_train)
train_preds = (train_preds > 0.5).astype(int)
cmatrix= confusion_matrix(y_true=y_train, y_pred=train_preds)
print(cmatrix)
precision = precision_score(y_train, train_preds)
recall = recall_score(y_train, train_preds)
f1 = f1_score(y_train, train_preds)
print("precision:",precision)
print("recall:",recall)
print("f1:",f1)
```

```python
predictions = model.predict(testData)
predictions = (predictions > 0.5).astype(int)
print(predictions)
```
